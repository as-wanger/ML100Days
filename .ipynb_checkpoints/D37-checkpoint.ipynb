## 作業
在鐵達尼資料集中，
透過進階補值方法 KNN，將鐵達尼號中的 Titanic_train
.csv 中的 age 的遺失值進行補值。    
* step1: 觀察 Age 和 Pclass 與 Sex 是否有關連性?
* step2: 如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。 

```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
```

## 讀入資料

```python
d_train = pd.read_csv("D35_Titanic_train.csv")
```

## 觀察 Age

step1：觀察 Age 和 Pclass 和 Sex 是否有關連性?
step2：如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。 

```python
#from sklearn.metrics.pairwise import nan_euclidean_distances
#display(pd.DataFrame(nan_euclidean_distances(d_train["Age","Pclass","Sex"])))
a = d_train.isnull().sum()/len(d_train)
a.sort_values(ascending=False)
```

```python
# 細看遺失資料的樣態，有177筆資料沒有年齡
missing_index = data.isnull()
data[missing_index.values==True]
```

```python
#同時比較三者，在二維圖上區分Age、Pclass再用hue區分Sex
sns.boxplot(x="Pclass", y="Age", hue="Sex", data=d_train)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
d_train['Sex']=le.fit_transform(d_train['Sex']) #f=0, m=1，離散轉連續型資料
d_train['Pclass']=le.fit_transform(d_train['Pclass'])#1~3變成0~2
data = d_train[["Sex","Pclass","Age"]]
data
```

```python
#資料點倆倆距離
from sklearn.metrics.pairwise import nan_euclidean_distances
data2 = pd.DataFrame(nan_euclidean_distances(data))
data2
#部分資料重疊如[0][5],[4][5]
```

```python
from sklearn.impute import KNNImputer
#設定k值
value_neighbors=1

imputer = KNNImputer(n_neighbors=value_neighbors)
df_filled = pd.DataFrame(imputer.fit_transform(data))
#補值前
print(data[missing_index.values==True])
#補值後
print(df_filled[missing_index.values==True])
```
