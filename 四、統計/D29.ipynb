### Q1:計算標準常態分配，小於1的機率有多大?
![](Q1.png)

```python
# library
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

#標準常態分配：平均值loc=0，標準差scale=1
x = np.linspace(-3,1,400)
y = stats.norm.cdf(x,0,1) #(x, loc=0, scale=1)
print("小於1的機率：%.4f" % (y[-1]))
plt.plot(x,y)
```

### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
![](Q2.png)


```python
## Q2
x = np.linspace(-1,1,400)
y = stats.norm.cdf(x,0,1)
print("大於1，小於 -1 的機率：%.4f" % (1+y[0]-y[-1]))
plt.plot(x,y)
```

### Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?

```python
#標準化：
Z = (6-2)/(4**0.5)
x = np.linspace(-Z,Z,400)
y = stats.norm.cdf(x,0,1)
print("計算 X小於 6 的機率：%.4f" % y[-1])
plt.plot(x,y)
```
